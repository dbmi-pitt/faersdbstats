```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import csv

terms = open("drugs-unmapped-June2019-lite.txt", 'r')
    
text = terms.readlines()

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(text)

true_k = 10
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print
print("\n")
print("Prediction")

output = open("cluster_10_size_10.txt","a")
for i in range(true_k):
    output.write('\n')
    output.write("Cluster %d:" % i),
    output.write('\n'),
    for ind in order_centroids[i, :20]:
        output.write(' %s' % terms[ind])
output.close()



true_k = 500
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :50]:
        print(' %s' % terms[ind]),
    print
print("\n")
print("Prediction")

output = open("cluster_500_size_50.txt","a")
for i in range(true_k):
    output.write('\n')
    output.write("Cluster %d:" % i),
    output.write('\n'),
    for ind in order_centroids[i, :50]:
        output.write(' %s' % terms[ind])
output.close()




true_k = 5000
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print
print("\n")
print("Prediction")

output = open("cluster_5000_size_10.txt","a")
for i in range(true_k):
    output.write('\n')
    output.write("Cluster %d:" % i),
    output.write('\n'),
    for ind in order_centroids[i, :10]:
        output.write(' %s' % terms[ind])
output.close()




true_k = 1000
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :25]:
        print(' %s' % terms[ind]),
    print
print("\n")
print("Prediction")

output = open("cluster_1000_size_25.txt","a")
for i in range(true_k):
    output.write('\n')
    output.write("Cluster %d:" % i),
    output.write('\n'),
    for ind in order_centroids[i, :25]:
        output.write(' %s' % terms[ind])
output.close()





true_k = 10000
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print
print("\n")
print("Prediction")

output = open("cluster_10000_size_10.txt","a")
for i in range(true_k):
    output.write('\n')
    output.write("Cluster %d:" % i),
    output.write('\n'),
    for ind in order_centroids[i, :10]:
        output.write(' %s' % terms[ind])
output.close()
```
